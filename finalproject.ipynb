```python
filtered_columns = ['name', 'categories'] + [col for col in dataframe.columns if col.startswith('location.')] + ['id']
dataframe_filtered = dataframe.loc[:, filtered_columns]

def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

dataframe_filtered['categories'] = dataframe_filtered.apply(get_category_type, axis=1)
dataframe_filtered.columns = [column.split('.')[-1] for column in dataframe_filtered.columns]

dataframe_filtered
```

```python
df = dataframe_filtered[['name','categories','distance','lat','lng']]
```

```python
df
```

```python
venues_map = folium.Map(location=[-12.066315, -77.001180], zoom_start=13)
for lat, lng, label in zip(df.lat, df.lng, df.categories):
    folium.features.CircleMarker([lat,lng], radius=5, color='red', popup=label, fill=True, fill_color='red', fill_opacity=0.6).add_to(venues_map)
venues_map
```

```python
df_binary = pd.get_dummies(df[['categories']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
df_binary['name'] = df['name'] 

# move neighborhood column to the first column
fixed_columns = [df_binary.columns[-1]] + list(df_binary.columns[:-1])
df_binary = df_binary[fixed_columns]

df_binary
```

```python
df_binary_grouped= df_binary.groupby('name').mean().reset_index()
df_binary_grouped
```

```python
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans
# set number of clusters
kclusters = 5

df_binary_grouped_clustering = df_binary_grouped.drop('name', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(df_binary_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```
